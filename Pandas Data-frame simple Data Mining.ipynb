# Pandas Dataframe | Data Mining

```python
#import packages
import numpy as np
import pandas as pd
```

```python
#import CSV as pandas dataframe
mining = pd.read_csv("supermarket_sales - Sheet1.csv")
#print head of the data
mining.head()
```

```python
#Print data information
mining.info()
```

### Descriptive Statistics

```python
#Print - Five Number Summary
mining.describe()
```

```python
#Print Descriptive Statistics Categorical Data
mining.describe(include=object)
```

### Summary Statisitcs

```python
#Print Value Counts of Categorical Variables
display(mining['Branch'].value_counts())
display(mining['City'].value_counts())
display(mining['Customer type'].value_counts())
display(mining['Gender'].value_counts())
display(mining['Product line'].value_counts())
display(mining['Payment'].value_counts())
```

```python
#Find the sum total of all expenditure (Total)
total = mining['Total'].cumsum()
total = total[999].round(2)
print("Total Value: $", total)

#Find the sum total of Gross Income 
gross = mining['gross income'].cumsum()
gross = gross[999].round(2)
print("Total Gross Income Value: $", gross)
```

```python
#groupby Gender.mean of column
display(mining.groupby('Gender')['Total'].mean())
#groupby Gender.median of column
display(mining.groupby('Gender')['Total'].median())

#groupby Ewallet.mean of column
display(mining.groupby('Customer type')['Total'].mean())
#groupby Ewallet.median of column
display(mining.groupby('Customer type')['Total'].median())

#groupby City.mean of column
display(mining.groupby('City')['Total'].mean())
#groupby City.median of column
display(mining.groupby('City')['Total'].median())

#groupby Payment.mean of column
display(mining.groupby('Payment')['Total'].mean())
#groupby Payment.median of column
display(mining.groupby('Payment')['Total'].median())
```

```python
#Multiple Groups and Summary Statistics. mean()
mining.groupby(['Gender','City','Payment'])[['Total','gross income']].mean()
```

```python
#Mulitple Groups and Summary Statistics. median()
mining.groupby(['Gender','City','Payment'])[['Total','gross income']].median()
```
